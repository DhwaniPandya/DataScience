```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df1 = pd.read_csv('2015.csv');
print(df1);
```

```python
df2 = pd.read_csv('countries of the world.csv');
print(df2);
```

```python
df3 = pd.merge(df1, df2, how='inner', left_on = 'Country',right_on='Country');
df3.to_excel('result.xlsx', sheet_name='Sheet1');
print(df3);
```

```python
df3.plot(x='Pop. Density (per sq. mi.)',y='Happiness Score',style='_')
df3 = pd.DataFrame(np.random.rand(10, 4), columns=['Happiness Rank', 'Pop. Density (per sq. mi.)', 'Net migration','Economy (GDP per Capita)'])
df3.plot.bar();
```
